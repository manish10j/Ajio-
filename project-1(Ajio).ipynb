```python
from selenium import webdriver
import numpy as np
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm
```

```python
path='D:\Manish\webdriver\chromedriver.exe'
driver=webdriver.Chrome(path)
driver.maximize_window()
```

```python

link='https://www.ajio.com/men-tshirts/c/830216014'
driver.get(link)
# raw_data=driver.find_elements_by_class_name('contentHolder')
```

```python
time.sleep(4)  # Allow 2 seconds for the web page to open
scroll_pause_time = 2# You can set your own pause time. My laptop is a bit slow so I use 1 sec
screen_height = driver.execute_script("return window.screen.height;")   # get the screen height of the web
print(screen_height)
i = 1
all_data = dict()

while True:
    # scroll one screen height each time
    total_screen = str(screen_height*i)
    print(total_screen)
    raw_data = driver.find_elements_by_class_name('contentHolder')
    for data in raw_data:
        if data not in all_data.keys():
            all_data[data]=data.text
    driver.execute_script(f"window.scrollTo(0, {total_screen} );")  
    i += 1
    print(i)
    
#     print(raw_data)
    time.sleep(scroll_pause_time)
    print(screen_height)
    # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
    scroll_height = driver.execute_script("return document.body.scrollHeight;")  
    # Break the loop when the height we need to scroll to is larger than the total scroll height
    print(scroll_height)
    if (screen_height) * i > scroll_height:
        break
    if len(all_data)>=500:
        break
    if i%30==0 and scroll_pause_time>=0.8:
        scroll_pause_time-=scroll_pause_time-0.2
```

```python
len(all_data)
type(all_data)
all_data

all_new_data = []

for value in all_data.values():
    all_new_data.append(value)
    
print(all_new_data)
```

```python
new_data = list()
brand, tShirt, discount_price, og_price, selling_price  = [list() for i in range(5)]

for i in range(0,len(all_new_data)):
    
    
    split_text = all_new_data[i].split('\n')
    
    brand.append(split_text[0])

    tShirt.append(split_text[1])

    price=split_text[2]
    #print(price)
    price_split=price.split('Rs. ')
    price_split_new=[price for price in price_split if price not in [""," "]]
    #print(price_split_new)
    if len(price_split_new)==3:
        discount_price.append(price_split_new[0])
        og_price.append(price_split_new[1])
    elif len(price_split_new)==2:
        discount_price.append(price_split_new[0])
        og_price.append(price_split_new[1])
    elif len(price_split_new)==1:
        discount_price.append("")
        og_price.append(price_split_new[0])

    if len(discount_price[i])==0: 
        selling_price.append(og_price[i])
    else:
        selling_price.append(discount_price[i])
        
new_data.append(list(zip(brand,tShirt,discount_price,og_price,selling_price)))
```

```python
print(new_data)
```

```python
all_df=[pd.DataFrame(data ,columns=['Brand','Product Type','Discount Price','Original Price','Selling Price']) for data in new_data]
```

```python
result=pd.concat(all_df,ignore_index=True)
result
```

```python
df_new=result.copy(deep=True)
```
